## CSV Files

```python
import pandas as pd
```

```python
pd.read_csv("iris.csv")
```

## Parameters of Pandas csv file
- nrows 
- usecols
- skiprows
- index_col 
- header
- prefix
- names

```python
pd.read_csv("iris.csv",usecols=[0,3]) ## usecols means access which column
```

```python
pd.read_csv("iris.csv",nrows=3,usecols=[0,2]) ## nrows mean access how many row 
```

```python
pd.read_csv("iris.csv",nrows=8)
```

```python
pd.read_csv("iris.csv",skiprows=[4],nrows=8) ## skiprows means which row skip 
```

```python
pd.read_csv("iris.csv",index_col='variety') ## index_col means which column set as a index also using indexing value like 4
```

```python
pd.read_csv("iris.csv",header=2) ## skip rows
```

```python
pd.read_csv("iris.csv",header=None)
```

```python
pd.read_csv("iris.csv",names=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Variety'])  ## rename the column name
```
