```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("adult.csv").fillna(method="ffill").head(11)
df
```

```python
length=df.iloc[:,0]
width=df.iloc[:,12]
x=np.arange(0,11)
y=np.arange(11,21)
print(length)
print(width)
```

```python
plt.plot(x,length,'go--',linewidth=2) ##  or (x,y,color='g',marker='o',linestyle='--')
plt.plot(x,width,'ro-.',linewidth=2,)
plt.title("Area")
plt.grid('k-')
plt.xlabel("x-axis")
plt.ylabel("length")
plt.legend(['Length','Width'],loc=2)
plt.show()
```
