## Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
arr1 = np.arange(0,48,3).reshape(2,8)
arr1
```

```python
sns.heatmap(arr1)
```

```python
df = sns.load_dataset("car_crashes")
df
```

```python
df=df.set_index("abbrev")
sns.heatmap(df)
```

```python
sns.heatmap(df,cmap="coolwarm")
```

```python
plt.figure(figsize=(20,14))
sns.heatmap(df,cmap="coolwarm",annot=True)
```

```python
arr2=np.arange(0,48,3).reshape(4,4)
arr2
```

```python
arr_annot=[]
for x in arr2:
    arr_annot.append([5*x[0],4* x[1],3* x[2],2* x[3]])
arr_annot
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d")
```

```python
plt.figure(figsize=(12,6))
kws={"fontsize":10,
   "fontstyle":"italic",
   "color":"k",
   "alpha":0.8,
   "rotation":"horizontal"}
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,linecolor="k")
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,cbar=False)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,xticklabels=False)
```

```python
plt.figure(figsize=(12,6))
ckws={
    "shrink":1,
    "extend":"min",
    "drawedges":True,
    "extendfrac":0.1
}
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,cbar_kws=ckws)
```

```python
plt.figure(figsize=(12,6))
ckws={
    "shrink":1,
    "extend":"min",
    "drawedges":True,
    "extendfrac":0.1
}
sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,cbar_kws=ckws,xticklabels=["A","B","C","D"], 
            yticklabels=["A_1","B_2","C_3","D_4"])
```

```python
plt.figure(figsize=(12,6))
ckws={
    "shrink":1,
    "extend":"min",
    "drawedges":True,
    "extendfrac":0.1
}
ax=sns.heatmap(arr2,cmap="coolwarm",annot=arr_annot,fmt="d",annot_kws=kws,linewidth=4,cbar_kws=ckws, 
               xticklabels=["A","B","C","D"], yticklabels=["A_1","B_2","C_3","D_4"])
ax.set(title="Array 2",xlabel="Columns",ylabel="Rows")
sns.set(font_scale=5)
```
