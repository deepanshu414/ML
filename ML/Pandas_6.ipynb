```python
df1=pd.read_csv("iris.csv").head(5)
df1
```

```python
df2=pd.read_csv("state.csv").head(5).replace({'Country Name':"Afghanistan"},'Setosa')
df2=df2.rename(columns={"Country Name":"variety"})
df2
```

```python
df3=pd.merge(df1,df2,on="variety")
df3
```

```python
pd.merge(df1,df2,on="variety",how="right")
```

```python
pd.merge(df1,df2,on="variety",how="outer",indicator=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,on="variety",suffixes=('petal.width','Country Code'))
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
display(df1,df2)
```

```python
df1.join(df2,how="right",rsuffix='_1',lsuffix="_2")
```
