```python
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset("tips") ## https://github.com/mwaskom/seaborn-data
tips
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],hist=False)
```

```python
sns.distplot(tips["total_bill"],kde=False)
```

```python
sns.distplot(tips["total_bill"],color="r",vertical=True)
sns.set()
```

```python
bins=np.arange(0,50)
sns.distplot(tips['total_bill'],bins)
```

```python
sns.distplot(tips['total_bill'],bins)
plt.xticks(bins)
```

```python
sns.distplot(tips["total_bill"],hist_kws={'color':"#DC143C",'edgecolor':'#aaff00','linewidth':1,'linestyle':'--','alpha':0.9})
```

```python
sns.distplot(tips["total_bill"],kde_kws={'color':"#8e00ce",'linewidth':1,'linestyle':'-.'})
```

```python
sns.distplot(tips["total_bill"],hist_kws={'color':"#DC143C",'edgecolor':'#aaff00','linewidth':1,'linestyle':'--','alpha':0.9} 
             ,kde_kws={'color':"#8e00ce",'linewidth':1,'linestyle':'-.'})
```
