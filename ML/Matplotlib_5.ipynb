```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
```

```python
df=pd.read_csv("adult.csv").fillna(method="ffill").head(11)
df
```

```python
length=df.iloc[:,0]
width=df.iloc[:,12].replace(80.0,20.0)
name=df.iloc[:,5]
pd.DataFrame({'Length':length,'Width':width,'Name':name})
```

```python
plt.pie(length,explode=[0,0,0.3,0,0,0,0,0,0,0,0],autopct="%0.1f%%",labels=name,radius=1.4,shadow=True,startangle=50)
style.use("ggplot")
plt.grid("k-")
plt.show()
```
