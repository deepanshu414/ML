## Nul values Handle

- na_values
- keep_default_na
- isnull()
- sum()
- notnull()
- dropna()
- how
- subset
- inplace
- fillna()

```python
import pandas as pd
```

```python
pd.read_csv("state.csv")
```

```python
pd.read_csv("state.csv",na_values='NaN').head(10)
```

```python
pd.read_csv("state.csv",na_values={'Country Code':'not available'})
```

```python
pd.read_csv("state.csv",keep_default_na=False)
```

```python
df = pd.read_csv("state.csv").head(15)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df.notnull()
```

```python
df.dropna() ## not return those value where null value exist any row 
```

```python
df.dropna(how="all")
```

```python
df.dropna(subset=['Country Code']) ## not retrun those value where country code column are use non value
```

```python
df1=pd.read_csv("state.csv").head(11)
df1
```

```python
df1.fillna(2) ## return 2 where null value are exist
```

```python
df1.fillna({'State Name':'Farah','Country Code':"AF",'Country Name':"Afghanistan"})
```

```python
df1.fillna(method="ffill") ## or 'pad' retrun previous value where null value
```

```python
df1.fillna(method="bfill")
```

```python
df1.fillna(0,limit=1) ## return 0 value where null value exist but only 1 value change by 0 in every column where null value
```
