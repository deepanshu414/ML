## Replace method of Pandas csv file

- replace
- to_replace
- regex

```python
import pandas as pd
```

```python
df1 = pd.read_csv("state.csv").head(11)
df1
```

```python
df1 = df1.fillna(method="ffill")
df1
```

```python
df1.replace(to_replace="Baghlan",value="Farah")
```

```python
df1.replace({'State Code':'DAY'},'DAYS') ## for particular column 
```

```python
df1.replace('[A-Za-z]',0,regex=True) ## replace 0 where string are available
```
