```python
import seaborn as sns 
```

```python
tips= sns.load_dataset("tips") ## https://github.com/mwaskom/seaborn-data
tips
```

```python
sns.lineplot(x="tip",y="total_bill",data=tips)
```

```python
sns.lineplot(x="size",y="tip",data=tips)
```

```python
sns.lineplot(x="size",y="total_bill",data=tips,hue="sex")
```

```python
sns.lineplot(x="size",y="total_bill",data=tips,hue="sex",style="sex",palette="Reds",dashes=False,markers=['o','<'])
sns.set(style="darkgrid")
```
