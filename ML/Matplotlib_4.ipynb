```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
```

```python
df=pd.read_csv("adult.csv").fillna(method="ffill").head(11)
df
```

```python
length=df.iloc[:,0]
width=df.iloc[:,12].replace(80.0,20.0)
name=df.iloc[:,5]
print(length)
print(width)
print(name)
```

```python
plt.scatter(length,width,c='b')
style.use("ggplot")
plt.title("Employee Data")
plt.xlabel("Workers")
plt.ylabel("Salary")
plt.grid("k-")
plt.show()
```
