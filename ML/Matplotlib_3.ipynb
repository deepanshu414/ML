```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
```

```python
df=pd.read_csv("adult.csv").fillna(method="ffill").head(11)
df
```

```python
length=df.iloc[:,0]
width=df.iloc[:,12].replace(80.0,20.0)
name=df.iloc[:,5]
pd.DataFrame({'Length':length,'Width':width,'Name':name})
```

```python
plt.bar(name,length,color='r')
plt.bar(name,width,color='k')
style.use("ggplot")
plt.grid('k-')
plt.show()
```
