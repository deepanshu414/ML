```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
image=img.imread("image.jpg")
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.title('Original Image')
plt.imshow(image)
```

```python
image
```

```python
image.ndim
```

```python
height, width, channels = image.shape
print(f"Original shape: {image.shape}") ## (matrices,rows,columns)
print(height)
print(width)
print(channels)
```

```python
# Convert to 2D array (grayscale)
if channels == 3:  # For RGB images
    grayscale = np.mean(image, axis=2)
elif channels == 4:  # For RGBA images
    grayscale = np.mean(image[:,:,:3], axis=2)

print(f"Grayscale shape: {grayscale.shape}")
```

```python
print(grayscale)
print(grayscale.ndim)
```

```python
plt.subplot(122)
plt.imshow(grayscale, cmap='gray')
plt.title('Grayscale Image')
```

```python
array_normalized = (grayscale - grayscale.min()) / (grayscale.max() - grayscale.min())
    
# Create a new figure
plt.figure(figsize=(10, 5))

# Display the normalized array as an image
plt.subplot(121)
plt.imshow(array_normalized, cmap='gray')
plt.title('Normalized Grayscale')

# Convert back to RGB
rgb_image = np.stack([array_normalized]*3, axis=2)
plt.subplot(122)
plt.imshow(rgb_image)
plt.title('Converted back to RGB')
```
