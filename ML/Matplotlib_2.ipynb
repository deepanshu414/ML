```python
import matplotlib.pyplot as plt
# from matplotlib import style
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("adult.csv").fillna(method="ffill").head(11)
df
```

```python
length=df.iloc[:,0]
y=np.arange(11,50,3)
print(length)
print(y)
```

```python
plt.axis([10,50,0,3])
plt.hist([y,length],bins=None,rwidth=0.8,color=['r','k'],histtype='bar')
plt.title("Data")
plt.xlabel("employee")
plt.ylabel("salary")
plt.legend("Workers",loc=2)
# style.use('ggplot')
plt.grid('k-')
plt.show()
```
